```python
import numpy as np
import matplotlib.pyplot as plt
import os

PATH = os.getcwd()
```

```python
def GanhoInversor(vi, R1, Rf):
        return (-vi * Rf/R1)

def GanhoNaoInversor(vi, R1, Rf):
    return (vi * ( 1 + Rf/R1))

def ResistorDeCompensacao(R1, Rf):
    return ((R1*Rf) / (R1+Rf))

def Medicao_Corrente_to_Tensao_NAOINVERSORA(Ii, Rs, R1, Rf):
    return (Ii*Rs*(1 + Rf/R1))
```

```python
Ii = np.linspace(0, 1, 1000)

Rs = 1
R1 = 20e6
Rf = 1e2

Vo = np.array(Medicao_Corrente_to_Tensao_NAOINVERSORA(np.cos(Ii), Rs, R1, Rf))

plt.plot(Ii, Vo)
plt.show()
```
